```python
#一個學生(student)的dictionary,格式如下:
#{'name':'value','chinese':78,'english':98,'math':67,'history':87,'health':76}
#請使用list幫我建立50個學生
```

```python
import random

# Function to generate a random score between 60 and 100
def generate_score():
    return random.randint(60, 100)

# List to store 50 students
students = []

# Generate 50 students
for i in range(50):
    student = {
        'name': f'student_{i+1}',
        'chinese': generate_score(),
        'english': generate_score(),
        'math': generate_score(),
        'history': generate_score(),
        'health': generate_score()
    }
    students.append(student)

# Print the list of students
for student in students:
    print(student)
```
